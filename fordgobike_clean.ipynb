```python
df_copy = df.copy()
```

```python
df.fillna({'start_station_id': 'kosong', 'start_station_name': 'kosong','end_station_id': 'missing','end_station_name': 'kosong'}, inplace=True)
```

```python
df = df.dropna(how='any', subset={'member_birth_year','member_gender'})
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df_copy.describe(include="all")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(df["duration_sec"])
```

```python
# persamaan 1
Q1 = df["duration_sec"].quantile(0.25)
Q3 = df["duration_sec"].quantile(0.75)

IQR  = Q3 - Q1

outlierQ3 = Q3 + 1.5 * IQR
```

```python
df = df[(df["duration_sec"] < outlierQ3)]
df["duration_sec"].describe()
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(df["duration_sec"])
```

```python
df.shape
```

```python
plt.figure(figsize = (15,15))
plt.boxplot(df["member_birth_year"])
```

```python
Q1 = df["member_birth_year"].quantile(0.25)
Q3 = df["member_birth_year"].quantile(0.75)
IQR  = Q3 - Q1

outlierQ1 = Q1 - 1.5 * IQR
```

```python
df = df[(df["member_birth_year"] > outlierQ1)]
df["member_birth_year"].describe()
```

```python
plt.figure(figsize = (15,15))
plt.boxplot(df["member_birth_year"])
```

```python
df.describe(include="all")
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
# pilih kolom numerik dari df
df_corr = df[[
    "duration_sec",
    "start_station_id",
    "start_station_latitude", "start_station_longitude",
    "end_station_id", "end_station_latitude", "end_station_longitude",
    "bike_id", "member_birth_year"
]]

# hitung matriks korelasi
corr_matrix = df_corr.corr()

# plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Heatmap Korelasi")
plt.show()
```
